TASK 2

```python
bases = np.arange(1, 11).reshape(-1, 1)
exponents = np.arange(1, 11)
numpy_matrix = np.power(bases, exponents)

column_names = ['base'] + [f'pow_{i}' for i in range(2, 11)]

df_powers = pd.DataFrame(numpy_matrix, columns=column_names)
print(df_powers)
```

TASK 3

```python
is_shared = np.shares_memory(numpy_matrix, df_powers.values)
print(f"Memory shared: {is_shared}")

if not is_shared:
    df_shared = pd.DataFrame(numpy_matrix, copy=False, columns=column_names)
    print("New shared DataFrame created.")
```

TASK 4

```python
print(f"Original Dtype: {df_powers.dtypes['pow_10']}")

df_downcasted = df_powers.apply(pd.to_numeric, downcast='integer')

print(f"Downcasted Dtype: {df_downcasted.dtypes['pow_10']}")
print(f"Memory reduction: {df_powers.memory_usage().sum() - df_downcasted.memory_usage().sum()} bytes")
```

TASK 5

```python
iris_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
iris = pd.read_csv(iris_url, header=None)

iris.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class']
```

TASK 6

```python
iris.columns = [col.replace(' ', '_') for col in iris.columns]

print(iris.head(10))
```

TASK 7

```python
print("Shallow memory usage:", iris.memory_usage().sum())
print("Deep memory usage:", iris.memory_usage(deep=True).sum())

iris_smaller = iris.copy()
numeric_cols = iris_smaller.select_dtypes(include=['float64']).columns
iris_smaller[numeric_cols] = iris_smaller[numeric_cols].astype('float32')

print("Reduced memory usage (float32):", iris_smaller.memory_usage(deep=True).sum())
```
