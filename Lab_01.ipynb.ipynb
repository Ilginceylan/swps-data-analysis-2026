TASK 4

```python
matrix = np.arange(1, 26).reshape(5, 5)
print("Full Matrix:\n", matrix)
print("Last value:", matrix[-1, -1])
print("Second column:", matrix[:, 1])
print("Third row:", matrix[2, :])
print("Main diagonal:", np.diag(matrix))
print("Sub-matrix (9,10,14,15):\n", matrix[1:3, 3:5])
```

```python
cube = np.arange(1, 28).reshape(3, 3, 3)
print("Elements where y=1 (vertical axis index 1):\n", cube[:, 1, :])
print("Values for z=0 and z=2 (depth indices - NumPy 0-indexed):\n", cube[:, :, [0, 2]])
```

TASK 6

```python
n = 10000

# Case A
start = time.time()
lst = list(range(n))
res_a = []
for x in lst:
    res_a.append(x * 2)
print(f"List for loop time: {time.time() - start:.6f}s")

# Case B
start = time.time()
res_b = [x * 2 for x in lst]
print(f"List comprehension time: {time.time() - start:.6f}s")

# Case C
start = time.time()
np_arr = np.arange(n)
res_c = np_arr * 2
print(f"NumPy array time: {time.time() - start:.6f}s")
```
